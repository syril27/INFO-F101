# Instructions conditionnelles et boucles

## Conditions

* `if / elif / else` : exécuté seulement si la condition est vraie

### Syntaxe complète

```python
if condition:
    instructions
elif condition:
    instructions
elif condition:
    instructions
else:
    instructions
```

### Indentation

* Convention : **4 espaces**

## Type booléen

* Deux valeurs possibles : `True`, `False`

Exemples :

```python
a = 3 < 5      # True
b = (10 == 2)  # False
```

## Opérateurs relationnels

* `==` : égal à
* `!=` : différent de
* `<`, `<=` : inférieur (ou égal)
* `>`, `>=` : supérieur (ou égal)

## Opérateurs logiques

* `and` : conjonction → `True` si les deux sont `True`
* `or` : disjonction → `True` si au moins un est `True`
* `not` : négation → inverse le résultat

Exemple :

```python
x = 5
print(x > 0 and x < 10)   # True
print(not(x < 0))         # True
```

## Logique booléenne

* Conjonction : ∧
* Disjonction : ∨
* Négation : ¬

### Lois de De Morgan

* ¬(A ∧ B) ⇔ (¬A) ∨ (¬B)
* ¬(A ∨ B) ⇔ (¬A) ∧ (¬B)

En Python :

```python
not (A and B)  ⇔  (not A) or (not B)
not (A or B)   ⇔  (not A) and (not B)
```

## Boucle `while`

* Répète un bloc tant que la condition est vraie

### Syntaxe

```python
while condition:
    instructions
```

### Exemple

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

## Boucle `for`

* La fonction `range(n)` génère une séquence de `0` à `n-1`.

### Syntaxe

```python
for x in sequence:
    instructions
```

* La variable `x` prend les valeurs successives de la séquence.
