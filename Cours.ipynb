# Instructions conditionnelles et boucles

## Conditions

* `if / elif / else` : exécuté seulement si la condition est vraie

### Syntaxe complète

```python
if condition:
    instructions
elif condition:
    instructions
elif condition:
    instructions
else:
    instructions
```

### Indentation

* Convention : **4 espaces**

## Type booléen

* Deux valeurs possibles : `True`, `False`

Exemples :

```python
a = 3 < 5      # True
b = (10 == 2)  # False
```

## Opérateurs relationnels

* `==` : égal à
* `!=` : différent de
* `<`, `<=` : inférieur (ou égal)
* `>`, `>=` : supérieur (ou égal)

## Opérateurs logiques

* `and` : conjonction → `True` si les deux sont `True`
* `or` : disjonction → `True` si au moins un est `True`
* `not` : négation → inverse le résultat

Exemple :

```python
x = 5
print(x > 0 and x < 10)   # True
print(not(x < 0))         # True
```

## Logique booléenne

* Conjonction : ∧
* Disjonction : ∨
* Négation : ¬

### Lois de De Morgan

* ¬(A ∧ B) ⇔ (¬A) ∨ (¬B)
* ¬(A ∨ B) ⇔ (¬A) ∧ (¬B)

En Python :

```python
not (A and B)  ⇔  (not A) or (not B)
not (A or B)   ⇔  (not A) and (not B)
```

## Boucle `while`

* Répète un bloc tant que la condition est vraie

### Syntaxe

```python
while condition:
    instructions
```

### Exemple

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

## Boucle `for`

* La fonction `range(n)` génère une séquence de `0` à `n-1`.

### Syntaxe

```python
for x in sequence:
    instructions
```

* La variable `x` prend les valeurs successives de la séquence.

# Fonctions en Python

## Fonctions prédéfinies

* `abs()` : renvoie la valeur absolue d’un nombre.
* `float()` : convertit une valeur en **nombre à virgule flottante**.
* `max()` : renvoie le **maximum** des valeurs en entrée.
* `math.sin()` et `math.cos()` : fonctions **trigonométriques**, disponibles dans le module `math`.
* `math.log()` : renvoie le **logarithme** du nombre en entrée.



## Définir une fonction

* Une **fonction** encapsule un **bloc de code réutilisable**.

### Syntaxe

```python
def nom_fonction(paramètres):
    instructions
```

### Exemple

```python
def saluer(nom):
    print(f"Bonjour, {nom} !")
```



## Séquence d’exécution

* Le programme s’exécute **ligne par ligne**.
* Une fonction est **définie une seule fois**, mais **appelée plusieurs fois**.



## Paramètres

* Les paramètres sont placés entre parenthèses, séparés par des virgules.
* L’appel doit **respecter l’ordre** et le **nombre** des paramètres.
* Lors de l’exécution, les **paramètres formels** deviennent des **variables locales**.



## Instruction `return`

* `return` permet de **renvoyer un résultat** à l’appelant.
* Dès qu’elle est exécutée, la fonction se termine.
* Sans `return`, la fonction renvoie **`None`**.

### Exemple

```python
def carre(n):
    return n * n

res = carre(5)
print(res)  # 25
```



## Confusion entre `print` et `return`

| Fonction  | Effet                              | Valeur renvoyée     |
| --------- | ---------------------------------- | ------------------- |
| `print()` | Affiche une valeur dans la console | `None`              |
| `return`  | Renvoie une valeur à l’appelant    | Valeur réutilisable |


## Fonctions booléennes

Une fonction peut **retourner un booléen** (`True` ou `False`), utile dans les conditions.

### Exemple

```python
def est_pair(n):
    return n % 2 == 0

print(est_pair(6))  # True
print(est_pair(7))  # False
```



## Fonctions sans paramètres

Une fonction peut être définie **sans paramètre** si elle n’a besoin d’aucune information externe.

### Exemple

```python
def afficher_titre():
    print("=== MON PROGRAMME ===")

afficher_titre()
```



## Annotation de type

L’**annotation de type** (ou *type hinting*) indique les types des arguments et du retour d’une fonction.

### Exemple

```python
def aire_rectangle(longueur: float, largeur: float) -> float:
    return longueur * largeur
```



## Variables locales et globales

* **Variable locale** : définie **dans une fonction**, n’existe que **pendant son exécution**.
* **Variable globale** : définie **en dehors d’une fonction**, accessible partout (sauf si masquée par une locale).

### Exemple

```python
x = 10  # globale

def f():
    x = 5  # locale
    print(x)  # 5

f()
print(x)  # 10
```



## Portée et visibilité des variables (LEGB)

Python suit le modèle **LEGB** :

| Niveau            | Description                                |
| ----------------- | ------------------------------------------ |
| **L (Local)**     | Variable définie dans la fonction en cours |
| **E (Enclosing)** | Variable d’une fonction englobante         |
| **G (Global)**    | Variable du module courant                 |
| **B (Built-in)**  | Fonctions et constantes intégrées à Python |
