## Opérateurs logiques

## Logique booléenne

* Conjonction : ∧
* Disjonction : ∨
* Négation : ¬

### Lois de De Morgan

* ¬(A ∧ B) ⇔ (¬A) ∨ (¬B)
* ¬(A ∨ B) ⇔ (¬A) ∧ (¬B)

En Python :

```python
not (A and B)  ⇔  (not A) or (not B)
not (A or B)   ⇔  (not A) and (not B)
```

# Fonctions en Python

## Fonctions prédéfinies

* `abs()` : renvoie la valeur absolue d’un nombre.
* `float()` : convertit une valeur en **nombre à virgule flottante**.
* `max()` : renvoie le **maximum** des valeurs en entrée.
* `math.sin()` et `math.cos()` : fonctions **trigonométriques**, disponibles dans le module `math`.
* `math.log()` : renvoie le **logarithme** du nombre en entrée.



## Définir une fonction

* Une **fonction** encapsule un **bloc de code réutilisable**.

### Syntaxe

```python
def nom_fonction(paramètres):
    instructions
```

### Exemple

```python
def saluer(nom):
    print(f"Bonjour, {nom} !")
```



## Séquence d’exécution

* Le programme s’exécute **ligne par ligne**.
* Une fonction est **définie une seule fois**, mais **appelée plusieurs fois**.



## Paramètres

* Les paramètres sont placés entre parenthèses, séparés par des virgules.
* L’appel doit **respecter l’ordre** et le **nombre** des paramètres.
* Lors de l’exécution, les **paramètres formels** deviennent des **variables locales**.



## Instruction `return`

* `return` permet de **renvoyer un résultat** à l’appelant.
* Dès qu’elle est exécutée, la fonction se termine.
* Sans `return`, la fonction renvoie **`None`**.

### Exemple

```python
def carre(n):
    return n * n

res = carre(5)
print(res)  # 25
```



## Confusion entre `print` et `return`

| Fonction  | Effet                              | Valeur renvoyée     |
| --------- | ---------------------------------- | ------------------- |
| `print()` | Affiche une valeur dans la console | `None`              |
| `return`  | Renvoie une valeur à l’appelant    | Valeur réutilisable |


## Fonctions booléennes

Une fonction peut **retourner un booléen** (`True` ou `False`), utile dans les conditions.

### Exemple

```python
def est_pair(n):
    return n % 2 == 0

print(est_pair(6))  # True
print(est_pair(7))  # False
```



## Fonctions sans paramètres



## Annotation de type

L’**annotation de type** (ou *type hinting*) indique les types des arguments et du retour d’une fonction.

### Exemple

```python
def aire_rectangle(longueur: float, largeur: float) -> float:
    return longueur * largeur
```



## Variables locales et globales

* **Variable locale** : définie **dans une fonction**, n’existe que **pendant son exécution**.
* **Variable globale** : définie **en dehors d’une fonction**, accessible partout (sauf si masquée par une locale).

### Exemple

```python
x = 10  # globale

def f():
    x = 5  # locale
    print(x)  # 5

f()
print(x)  # 10
```



## Portée et visibilité des variables (LEGB)

Python suit le modèle **LEGB** :

| Niveau            | Description                                |
| ----------------- | ------------------------------------------ |
| **L (Local)**     | Variable définie dans la fonction en cours |
| **E (Enclosing)** | Variable d’une fonction englobante         |
| **G (Global)**    | Variable du module courant                 |
| **B (Built-in)**  | Fonctions et constantes intégrées à Python |

# Séquences en Python

## Chaînes de caractères (`str`)

* Une **chaîne** (string) est une **séquence de caractères**.
* Type Python : `str`.
* Les **chaînes sont immuables** → on ne peut pas les modifier.
* Indexation, **tranches** (slicing) et **méthodes intégrées** disponibles.

### Exemple

```python
s = "Python"
print(s[0])      # 'P'
print(s[1:4])    # 'yth'
print(len(s))    # 6
print(s.upper()) # 'PYTHON'
```



## Indices

* On accède aux éléments d’une séquence avec `sequence[index]`.
* Les indices vont de **0 à n−1**, où `n` est la longueur de la séquence.
* Les **indices négatifs** existent :

  * `-1` → dernier élément
  * `-2` → avant-dernier, etc.


## Tuples et listes

* **Liste (`list`)** : séquence **muable**, entre crochets `[ ]`.
* **Tuple (`tuple`)** : séquence **immuable**, entre parenthèses `( )`.

### Exemple

```python
L = [1, 2, 3]
print(type(L))       # <class 'list'>
L.append(4)          # [1, 2, 3, 4]
L[0] = -1            # [-1, 2, 3, 4]

T = (10, 20, 30)
print(type(T))       # <class 'tuple'>
# T[0] = 100 → Erreur : tuple immuable

print(len(T))        # 3
print(L[1])          # 2
```

 **Tuples** pour données fixes, **listes** pour données dynamiques.



## Opérateur `in`

Deux usages principaux :

1. **Itération** dans une boucle
2. **Test d’appartenance**

### Exemple

```python
# 1. Boucle
for lettre in "Python":
    print(lettre)

# 2. Test d’appartenance
voyelles = ['a', 'e', 'i', 'o', 'u', 'y']
print('e' in voyelles)  # True
print('z' in voyelles)  # False
```



## Slicing (Tranches)

Permet d’extraire une **sous-partie** d’une séquence.

### Syntaxe

```python
sequence[start:stop:step]
```

### Exemple

```python
s = [0, 1, 2, 3, 4, 5]
print(s[1:4])   # [1, 2, 3]
print(s[:3])    # [0, 1, 2]
print(s[::2])   # [0, 2, 4]
print(s[::-1])  # [5, 4, 3, 2, 1, 0]
```

* La **borne supérieure n’est pas incluse**.
* Fonctionne avec les **listes**, **tuples** et **chaînes**.
* Les tranches créent **de nouvelles séquences**.



## Listes et assignations

Lorsqu’on écrit :

```python
A = [1, 2, 3, 4]
B = A
B[0] = -1
print(A)  # [-1, 2, 3, 4]
```

 `A` et `B` pointent vers **le même objet en mémoire**.
La variable ne copie pas la liste, elle **copie la référence (pointeur)**.



## Identité et égalité

| Opérateur | Teste...                            | Exemple  | Résultat          |
| --------- | ----------------------------------- | -------- | ----------------- |
| `==`      | L’**égalité de contenu**            | `a == b` | `True`            |
| `is`      | L’**identité mémoire** (même objet) | `a is b` | `False` (souvent) |

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)  # True  → même contenu
print(a is b)  # False → objets distincts
print(a is c)  # True  → même objet
```



## Paramètres de fonctions : muables ou immuables

* Les **objets muables** (ex : listes) peuvent être **modifiés** dans une fonction.
* Les **objets immuables** (ex : entiers, tuples, chaînes) **ne le peuvent pas**.

### Exemple

```python
def modifie(liste):
    liste.append(4)

a = [1, 2, 3]
modifie(a)
print(a)  # [1, 2, 3, 4]


def modifie_str(s):
    s = s + "!"

b = "bonjour"
modifie_str(b)
print(b)  # "bonjour"
```

 Les objets **muables** sont **passés par référence**.



## Séquences de séquences

Les **listes** et **tuples** peuvent contenir **des éléments de types variés**, y compris d’autres listes ou tuples.

### Exemple

```python
L = ['bonjour', 14, 3.14, True]
T = ('bonjour', 14, 3.14, True)

L2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
L3 = [(1, 2), (3, 4), (5, 6)]

print(type(L3))        # <class 'list'>
print(type(L3[0]))     # <class 'tuple'>
print(type(L3[0][0]))  # <class 'int'>
```

## Compréhensions de liste

Permettent de créer des listes **en une seule ligne**.

### Syntaxe

```python
[expression for variable in iterable if condition]
```

### Exemples

```python
# Carrés de 0 à 9
carres = [x**2 for x in range(10)]

# Nombres pairs de 0 à 19
pairs = [x for x in range(20) if x % 2 == 0]
```
