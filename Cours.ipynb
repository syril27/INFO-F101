# Fonctions en Python

## Fonctions prédéfinies

## Définir une fonction

* Une **fonction** encapsule un **bloc de code réutilisable**.

### Syntaxe

```python
def nom_fonction(paramètres):
    instructions
```


## Séquence d’exécution

* Le programme s’exécute **ligne par ligne**.
* Une fonction est **définie une seule fois**, mais **appelée plusieurs fois**.



## Paramètres

* Les paramètres sont placés entre parenthèses, séparés par des virgules.
* L’appel doit **respecter l’ordre** et le **nombre** des paramètres.

## Fonction `histogramme()`

Renvoie un **dictionnaire** des lettres d’une chaîne avec leur **fréquence d’apparition**.

### Exemple

```python
def histogramme(s):
    """Renvoie le dictionnaire des lettres dans s avec leur fréquence."""
    d = {}  # dictionnaire vide

print(histogramme("bonjour"))

```

# Récursivité en Python



## Qu’est-ce que la récursivité ?

* La **récursivité** est une **technique de programmation** où **une fonction s’appelle elle-même**.

* Elle est utilisée pour résoudre un problème en le divisant en **sous-problèmes similaires**.

### Une fonction récursive comporte :

* Un ou plusieurs **cas de base** (ou conditions d’arrêt).
* Un ou plusieurs **appels récursifs**.

##  Multiplication par addition (récursive)

$$
a \times b = a + a \times (b - 1)
$$


```python
def multiplication_recursive(a, b):
    if b == 0:
        return 0
    else:
        return a + multiplication_recursive(a, b - 1)
```


## Définition récursive de la factorielle

$$
n! =
\begin{cases}
1 & \text{si } n = 0 \\
n \times (n-1)! & \text{si } n > 0
\end{cases}
$$



### Implémentation :

```python
def factorielle(n):
    if n == 0:
        return 1
    else:
        return n * factorielle(n - 1)
```

**Remarques :**

* Chaque appel crée un **nouveau contexte** (valeur propre de `n`).

* Les appels sont **empilés**, puis **dépilés** à la fin.

* La taille de la pile dépend de `n`.

<img src="https://i.ytimg.com/vi/l-r1JffTvHA/maxresdefault.jpg" width=60%>

##  Appel récursif unique

Un **appel récursif unique** signifie qu’il n’y a **qu’un seul appel** à la fonction dans son corps.

```python
def rec(n):
    if n > 0:
        print('a')
        rec(n - 1)
        print('b')
```

→ La fonction s’appelle **une seule fois par exécution**.



##  Récursivité terminale

Une **récursion terminale** a lieu lorsque **l’appel récursif est la dernière instruction** de la fonction.
Cela permet une **meilleure gestion de la pile** (optimisation possible).

```python
def fact(n, acc=1):
    if n == 0:
        return acc
    return fact(n - 1, acc * n)  # appel récursif terminal
```


##  Récursivité et induction

La **récursivité** en programmation est proche de **l’induction mathématique** :

| Étape | Récursivité    | Induction                  |
| ----- | -------------- | -------------------------- |
| 1     | Cas de base    | Initialisation             |
| 2     | Appel récursif | Passage à l’étape suivante |

 En induction, on prouve une propriété pour `n = 0`, puis on montre que **si elle est vraie pour n, elle l’est aussi pour n + 1**.


##  Reconnaître un palindrome

Un **palindrome** est un mot qui se lit **de la même manière dans les deux sens**.

### Méthode récursive :

1. Cas de base : chaîne vide ou à un seul caractère → palindrome.
2. Si les extrémités sont égales → vérifier le mot sans les extrêmes.
3. Sinon → ce n’est pas un palindrome.

```python
def est_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return est_palindrome(s[1:-1])
```

Exemples :

```python
est_palindrome("radar")  # True
est_palindrome("python") # False
```



##  Tours de Hanoï

### Règles :

* Un seul disque déplacé à la fois.
* Aucun disque plus grand ne peut être posé sur un plus petit.

### Principe récursif :

1. **Cas de base** : 1 disque → déplacement direct.
2. **Sinon** :

   * Déplacer `n-1` disques du piquet source vers l’auxiliaire.
   * Déplacer le disque `n` vers la cible.
   * Déplacer les `n-1` disques de l’auxiliaire vers la cible.

### Code :

```python
def hanoi(n, source, cible, auxiliaire):
    if n == 1:
        print(f"Déplacer disque 1 de {source} à {cible}")
    else:
        hanoi(n - 1, source, auxiliaire, cible)
        print(f"Déplacer disque {n} de {source} à {cible}")
        hanoi(n - 1, auxiliaire, cible, source)

hanoi(3, 'A', 'B', 'C')
```

# Tris

## Principe général

Trier une collection d’objets consiste à **réorganiser ses éléments selon un ordre donné**.

### Pourquoi trier ?

* Facilite la **recherche** (ex. : recherche dichotomique)
* Améliore les **performances d’autres algorithmes** (fusion, suppression de doublons, etc.)
* Présent dans de nombreux domaines : **bases de données**, **compression**, **affichage**, **moteurs de recherche**, etc.

### Historique

* **Années 1950** : tri de cartes perforées
* **1960–1970** : apparition des grands algorithmes (bulle, insertion, Shell, HeapSort, QuickSort)
* **1960** : Tony Hoare invente le **tri rapide (QuickSort)**, encore largement utilisé aujourd’hui



## Tri à bulles

**Principe :** comparer chaque paire d’éléments adjacents et les échanger si nécessaire.
Le plus grand élément « remonte » à chaque passage, comme une bulle.

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20240925153535/bubble-sort-1.webp" width="500">

### Implémentation

```python
def tri_bulle(tab):
    n = len(tab)
    for i in range(n):
        for j in range(0, n - i - 1):
            if tab[j] > tab[j + 1]:
                tab[j], tab[j + 1] = tab[j + 1], tab[j]
```

À chaque itération, le plus grand élément est placé à la fin : la zone à parcourir diminue progressivement.


## Tri par insertion

**Principe :** insérer chaque élément à sa place dans la partie déjà triée de la liste.

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20240802210251/Insertion-sorting.png" width="500">

### Implémentation

```python
def tri_insertion(tab):
    for i in range(1, len(tab)):
        cle = tab[i]
        j = i - 1
        while j >= 0 and tab[j] > cle:
            tab[j + 1] = tab[j]
            j -= 1
        tab[j + 1] = cle
```

Chaque nouvel élément est inséré à la bonne position parmi ceux déjà triés.



## Tri rapide (QuickSort)

**Principe récursif :**

1. Choisir un **pivot** (souvent le premier élément)
2. Diviser la liste :

   * `left` : éléments ≤ pivot
   * `right` : éléments > pivot
3. Trier récursivement `left` et `right`
4. Reconstituer la liste triée :
   `quicksort(left) + [pivot] + quicksort(right)`


<img src="https://media.geeksforgeeks.org/wp-content/uploads/20231219164812/Quick-Sort-Algorithm.png" width="500">


### Implémentation

```python
def quicksort(lst):
    if len(lst) <= 1:
        return lst
    pivot = lst[0]
    left = [x for x in lst[1:] if x <= pivot]
    right = [x for x in lst[1:] if x > pivot]
    return quicksort(left) + [pivot] + quicksort(right)
```

Le **QuickSort** est un **algorithme récursif et très efficace**, souvent plus rapide que le tri à bulles ou à insertion.

# Quantificateurs universels et existentiels

En logique :

* **$\forall x, P(x)$** : vrai si $P(x)$ est vraie pour toutes les valeurs de $x$.  
* **$\exists x, P(x)$** : vrai si $P(x)$ est vraie pour au moins une valeur de $x$.



## Quantificateur universel en Python : `all`

```python
all(premier(x) for x in L)
```

### Exemple : vérifier qu’une liste est triée

Formule logique :

$\forall j \in \{0, \ldots, n-2\},\; L[j] \le L[j+1]$


En Python :

```python
assert all(L[j] <= L[j+1] for j in range(len(L)-1)), "La liste n’est pas triée !"
```

Ou sous forme de fonction :

```python
def est_trie(L):
    return all(L[j] <= L[j+1] for j in range(len(L)-1))
```

Formule :
$\exists x \in L,\; P(x)$


Python :


# Invariants de boucle

## Définition

Un invariant est une propriété logique qui reste **toujours vraie** :

* avant chaque itération
* pendant chaque itération
* après la boucle

## Rôle

Il sert à prouver la correction d’un algorithme.



# Exemples importants

## 1. Recherche linéaire

Invariant :
« L’élément cherché n’est pas dans la partie déjà parcourue. »

```python
def recherche(L, e):
    for i in range(len(L)):
        assert all(L[j] != e for j in range(i))
        if L[i] == e:
            return i
```

Invariant :
« La sous-liste de gauche (jusqu’à l’indice i-1) est triée. »

```python
def tri_insertion(tab):
    for i in range(1, len(tab)):
        assert est_trie(tab[:i])
        cle = tab[i]
        j = i - 1
        while j >= 0 and tab[j] > cle:
            tab[j+1] = tab[j]
            j -= 1
        tab[j+1] = cle
```

# Complexité

## `Efficacité` des programmes

* L’efficacité reste importante malgré l’augmentation de la puissance des ordinateurs.

* Les volumes de données peuvent être très grands, ce qui rend les solutions naïves insuffisantes.

* L’efficacité se divise en efficacité en temps et en espace.

* Ici, on se concentre uniquement sur l’efficacité en temps.



## Complexité en `temps`

Plusieurs possibilités pour évaluer la complexité en temps :

1. Mesurer le temps d’exécution avec un minuteur.
2. Compter le nombre d’opérations effectuées.
3. Utiliser une notion d’ordre de croissance.

### Exemple : mesure du temps d’exécution

```python
import time

def c_to_f(c):
    return c * 9 / 5 + 32

t0 = time.clock()
c_to_f(100000)
t1 = time.clock() - t0
print("t =", t1, "s")
```

* Le but est d’évaluer des **algorithmes**.
* Le temps d’exécution varie selon la machine, l’implémentation, les entrées.
* Aucun lien clair entre taille de l’entrée et durée.



## Compter les opérations

* On suppose que certaines opérations prennent un temps constant : arithmétiques, comparaisons, assignations.

Exemple :

```python
x = 0      # 1 assignation
for i in range(n):   # n itérations
    x = x + 1        # 1 addition + 1 assignation par itération
```

Nombre total :
**1 + 3n**

* Dépend de l’algorithme, pas de la machine.
* Permet d’obtenir une relation entre la taille de l’entrée et le nombre d’opérations.


## Comportement `asymptotique`

* On évalue l’efficacité pour des **entrées très grandes**.
* On exprime la croissance du temps d’exécution.
* On obtient une **borne supérieure**, pas une valeur exacte.



## `Meilleur`, `pire` et cas `moyen`

Exemple : recherche linéaire

```python
def recherche(L, e):
    for x in L:
        if x == e:
            return True
    return False
```

* **Meilleur cas** : l’élément est au début → temps constant.
* **Pire cas** : l’élément est absent → temps proportionnel à n.
* **Cas moyen** : l’élément est vers le milieu → n/2.

En théorie, on utilise souvent le **pire cas**.



## Borne supérieure asymptotique

Définition :

Soient deux fonctions f et g.
On dit que :

$f(n) = O(g(n))$

s'il existe $N > 0$ et $c > 0$ tels que :

$$
\forall n > N,\quad f(n) < c \cdot g(n)
$$


Interprétation : f est **bornée supérieurement** par g à une constante multiplicative près.



## Application

Dans notre contexte, f(n) représente **le nombre d’opérations** effectuées.

La notation O(g(n)) :

* décrit une *borne supérieure asymptotique* (pire cas),
* ignore les constantes,
* exprime la croissance selon la taille de l’entrée.



## Exemples

### `Boucle` simple

```python
def fact_iter(n):
    answer = 1
    while n > 1:
        answer *= n
        n -= 1
    return answer
```

→ Complexité : **O(n)**


### Recherche `linéaire`

Même code qu’au-dessus.
→ Complexité : **O(n)**



## Deux boucles `imbriquées`

```python
for i in range(n):
    for j in range(n):
        print('a')
```

→ Complexité : **O(n²)**



## Simplifications

* O(n² + 2n + 2) = O(n²)
* O(n² + 100000n + 31000) = O(n²)
* O(log n + n + 4) = O(n)
* O(0.0001 n log n + 300n) = O(n log n)
* O(2n³ + 3n) = O(n³)



## Lois de `composition`

### `Séquence`

```python
for i in range(N): print('a')
for j in range(M): print('b')
```

→ Complexité : **O(N + M)**



### `Imbrication`

```python
for i in range(N):
    for j in range(M):
        print("ab")
```

→ Complexité : **O(N × M)**



### `Branchement`

```python
if condition:
    for i in range(N):
        print('a')
else:
    for j in range(M):
        print('b')
```

→ Complexité : **O(max(N, M))**



## Classes de `complexité`

* `O(1)` : constante
* `O(log n)` : logarithmique
* `O(n)` : linéaire
* `O(n log n)` : log-linéaire
* `O(nᶜ)` : polynomiale
* `O(cⁿ)` : exponentielle

## Exemples de croissance

| Fonction | n=10 | n=100   | n=1000   | n=1 000 000  |
| -------- | ---- | ------- | -------- | ------------ |
| 1        | 1    | 1       | 1        | 1            |
| log n    | 1    | 2       | 3        | 6            |
| n        | 10   | 100     | 1000     | 1e6          |
| n log n  | 10   | 200     | 3000     | 6e6          |

## Exercice

Un ensemble $A$ est inclus dans un ensemble $B$ si tous les éléments de $A$ se trouvent également dans $B$.  
On dit alors que $A$ est un sous-ensemble de $B$.

### Implémentation

```python
def sous_ensemble(L1, L2):
    for e1 in L1:
        matched = False
        for e2 in L2:
            if e1 == e2:
                matched = True
                break
        if not matched:
            return False
    return True
````

* On parcourt chaque élément de `L1`.
* Pour chacun, on parcourt `L2` afin de vérifier sa présence.
* Complexité : $O(n \cdot m)$
  avec $n = |L1|$ et $m = |L2|$.

## Aspects de l’informatique théorique

* **Analyse d’algorithmes**
  Raisonner sur un algorithme pour prédire le temps nécessaire selon la taille de l’entrée.

* **Conception d’algorithmes**
  Relier les choix de conception à l’efficacité temporelle.

* **Théorie de la complexité**
  Étudier les limites fondamentales du temps nécessaire pour résoudre un problème.

## Ordres de croissance : rappel

Objectifs :

* Évaluer l’efficacité pour des entrées très grandes.
* Exprimer la croissance du temps d’exécution en fonction de la taille de l’entrée.
* Établir une borne supérieure aussi précise que possible.
* Parler d’**ordre de croissance**, pas de valeur exacte.
* Se concentrer sur les facteurs dominants.

## Classes de complexité

### Complexité constante

* Indépendante de la taille de l’entrée.
* Peu d’algorithmes complets dans cette classe.
* Certaines parties d’algorithmes peuvent être en $O(1)$.
* Peut contenir des boucles ou récursions si leur nombre est indépendant de l’entrée.

### Complexité logarithmique

* Croissance proportionnelle à $\log(n)$.
* $\log_2(n)$ correspond approximativement au nombre de bits nécessaires pour écrire $n$ en binaire.
* Exemple typique : **recherche dichotomique**.

## Recherche dichotomique

On cherche un élément $e$ dans une liste triée de $n$ éléments.

Principe :

1. Choisir l’indice $i = n // 2$.
2. Tester si $L[i] = e$.
3. Sinon, choisir la moitié pertinente.
4. Répéter récursivement.

### Implémentation récursive

```python
def recherche_dichotomique(L, e):
    if L == []:
        return False
    elif len(L) == 1:
        return L[0] == e
    else:
        moitie = len(L) // 2
        if L[moitie] > e:
            return recherche_dichotomique(L[:moitie], e)
        else:
            return recherche_dichotomique(L[moitie:], e)
```

* Complexité théorique : $O(\log n)$.
* Attention : les découpages `L[:]` copient la liste.
* Coût réel : $O(n)$ par appel.

## Logarithmes

Considérons la boucle suivante :

```python
while n > 1:
    n = n // 2
```

* Le nombre d’itérations correspond au nombre de divisions successives par $2$.
* Ce nombre vaut $\lfloor \log_2(n) \rfloor$.
* Complexité : $O(\log n)$.

## Complexité linéaire

Exemples :

* Recherche séquentielle dans une liste.
* Boucles simples parcourant une structure une seule fois.

Complexité : $O(n)$.

## Tris

### Tri à bulles (Bubble Sort)

Principe :

* Comparer chaque paire d’éléments adjacents.
* Échanger si nécessaire.
* Répéter jusqu’à ce que la liste soit triée.

```python
def tri_bulle(tab):
    n = len(tab)
    for i in range(n):
        for j in range(0, n - i - 1):
            if tab[j] > tab[j+1]:
                tab[j], tab[j+1] = tab[j+1], tab[j]
```

* Complexité : $O(n^2)$.

### Tri par insertion

Principe :

* Insérer chaque élément à sa place dans la partie déjà triée.

```python
def tri_insertion(tab):
    for i in range(1, len(tab)):
        cle = tab[i]
        j = i - 1
        while j >= 0 and tab[j] > cle:
            tab[j+1] = tab[j]
            j -= 1
        tab[j+1] = cle
```

* Complexité : $O(n^2)$.
* Très efficace pour des listes presque triées.

### Tri rapide (Quick Sort)

Principe :

* Choisir un pivot.
* Répartir les éléments plus petits et plus grands.
* Trier récursivement.

```python
def quicksort(tab):
    if len(tab) <= 1:
        return tab[:]
    pivot = tab[0]
    gauche = [x for x in tab[1:] if x <= pivot]
    droite = [x for x in tab[1:] if x > pivot]
    return quicksort(gauche) + [pivot] + quicksort(droite)
```

* Complexité moyenne : $O(n \log n)$.
* Pire cas : $O(n^2)$.

### Fusion de deux listes triées

```python
def fusion(g, d):
    resultat = []
    i = j = 0
    while i < len(g) and j < len(d):
        if g[i] < d[j]:
            resultat.append(g[i])
            i += 1
        else:
            resultat.append(d[j])
            j += 1
    return resultat + g[i:] + d[j:]
```

* Complexité : $O(n)$.

### Tri fusion (Merge Sort)

Principe :

* Diviser récursivement la liste.
* Trier chaque moitié.
* Fusionner.

```python
def merge_sort(tab):
    if len(tab) <= 1:
        return tab
    milieu = len(tab) // 2
    gauche = merge_sort(tab[:milieu])
    droite = merge_sort(tab[milieu:])
    return fusion(gauche, droite)
```

* Complexité : $O(n \log n)$.

## Complexité polynomiale

### Produit de matrices $n \times n$

```python
def produit_matrices(A, B):
    n = len(A)
    C = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

* Trois boucles imbriquées.
* Complexité : $O(n^3)$.


## Complexité exponentielle

### Sous-ensembles

* L’ensemble des parties de $B$ est noté $P(B)$.
* On a $|P(B)| = 2^{|B|}$.

Exemple :

* L’ensemble ${1,2,3,4}$ possède $2^4 = 16$ sous-ensembles.

### Génération récursive des sous-ensembles

```python
def sous_ensembles(L):
    if len(L) == 0:
        return [[]]
    petits = sous_ensembles(L[:-1])
    extra = [L[-1]]
    nouveaux = []
    for petit in petits:
        nouveaux.append(petit + extra)
    return petits + nouveaux
```

* Complexité : $O(2^n)$.

### Tours de Hanoï

Problème :

* Déplacer $n$ disques entre trois tours.
* Un seul disque à la fois.
* Aucun disque ne peut être posé sur un disque plus petit.

Relation de récurrence :

* $t_n = 2t_{n-1} + 1$
* Solution : $t_n = 2^n - 1$

Conclusion :

* Complexité : $O(2^n)$.

## Classes de complexité (récapitulatif)

* $O(1)$ : indépendant de la taille.

* $O(\log n)$ : dichotomie.
* $O(n)$ : itération simple.
* $O(n \log n)$ : diviser pour régner.
* $O(n^c)$ : boucles imbriquées.
* $O(c^n)$ : récursivité exponentielle (sous-ensembles, Tours de Hanoï).



## Complexité des operations en Python    



### List
<img src="https://drive.google.com/uc?export=view&id=1UJNeKXhRh1l-0YyOsfZ7bn_mJwsUCkRp" alt="Image Google Drive" width="400">

###Doubly linked list
<img src="https://drive.google.com/uc?export=view&id=19RbinmwPi5jTpUXF8JVcNmJfLkB1bLQ4" alt="Image Google Drive" width="400">


### Set
<img src="https://drive.google.com/uc?export=view&id=1vn8KpoU9l542KDKWbZPLxTX-9zubxqjP" alt="Image Google Drive" width="400">

### Dic
<img src="https://drive.google.com/uc?export=view&id=1yh231852CR91tGPS97yYzes24h41oGT6" alt="Image Google Drive" width="400">
