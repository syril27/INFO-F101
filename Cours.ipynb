Permet d‚Äôextraire une **sous-partie** d‚Äôune s√©quence.

### Syntaxe

```python
sequence[start:stop:step]
```

### Exemple

```python
s = [0, 1, 2, 3, 4, 5]
print(s[1:4])   # [1, 2, 3]
print(s[:3])    # [0, 1, 2]
print(s[::2])   # [0, 2, 4]
print(s[::-1])  # [5, 4, 3, 2, 1, 0]
```

## Identit√© et √©galit√©

| Op√©rateur | Teste...                            | Exemple  | R√©sultat          |
| --------- | ----------------------------------- | -------- | ----------------- |
| `==`      | L‚Äô**√©galit√© de contenu**            | `a == b` | `True`            |

* Les **objets muables** (ex : listes) peuvent √™tre **modifi√©s** dans une fonction.
* Les **objets immuables** (ex : entiers, tuples, cha√Ænes) **ne le peuvent pas**.

## Fichiers

### √âcrire dans un fichier

La fonction `open()` permet d‚Äô**ouvrir un fichier** avec un **nom** et un **mode** :

| Mode  | Signification                         |
| ----- | ------------------------------------- |
| `"w"` | √©criture (√©crase le contenu existant) |
| `"r"` | lecture                               |
| `"a"` | ajout √† la fin du fichier             |

### Exemple

```python
f = open("notes.txt", "w")
f.write("Bonjour\n")
f.write("Ligne suivante\n")
f.close()
```

 La m√©thode `close()` **ferme le fichier** et termine les op√©rations d‚Äô√©criture.

### Encodage

On peut pr√©ciser l‚Äôencodage (souvent `utf-8`) :

```python
f = open("res.txt", "w", encoding="utf-8")
```



## Lire un fichier ligne par ligne

Les m√©thodes **`read()`**, **`readline()`** et **`readlines()`** permettent de lire le contenu.

### Exemple avec `for` sur le fichier

```python
f = open("notes.txt", "r")
for ligne in f:
    print(ligne.strip())
f.close()
```

### Exemple avec `readlines()`

```python
f = open("notes.txt", "r")
lignes = f.readlines()
for l in lignes:
    print(l.upper())
f.close()
```

Chaque ligne lue contient un **symbole de fin de ligne** `\n` √† la fin.



## `with ... as`

La structure `with ... as` permet une **gestion automatique** de l‚Äôouverture et de la fermeture du fichier (pas besoin de `close()`).

### Exemple

```python
with open("notes.txt", "r") as f:
    for ligne in f:
        print(ligne.strip())
```

üëâ Le fichier est **ferm√© automatiquement** √† la fin du bloc.
